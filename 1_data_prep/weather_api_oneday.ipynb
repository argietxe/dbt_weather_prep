```python
# import the api_key from .env
import os
from dotenv import load_dotenv
load_dotenv()
api_key = os.getenv('WEATHER_API')
```

```python
# call api url
import json
import requests
import pprint # https://docs.python.org/3/library/pprint.html
import pandas as pd

api_url = f'http://api.weatherapi.com/v1/history.json?key={api_key}&q=Berlin&dt=2023-05-24'

response = requests.request("GET", api_url)   #is a json object

result = response.json()     #for it to make it as text form. 
```

```python
# saving as json 
with open('weather_berlin.json', mode='w') as f:
    f.write(response.text)
```
